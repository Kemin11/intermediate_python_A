### Handling missing value

```python
import numpy as np
```

```python
df = pd.DataFrame({'col_A' : [1,2,3,1,np.nan,np.nan,5],
                   'col_B' : [3,7,np.nan,9,None,5,8],
                   'col_C' : ['a','?', 'x', 'y','--',np.nan,'r'],
                   'col_D' : [True, True, np.nan, None, False, True, False]})
df
```

```python
#membuat csv baru
df.to_csv('newDF.csv', index=False)
```

```python
df.shape
```

```python
df.info()
#keterangan : non-null menunjukkan data yang tidak hilang
```

#### Find missing value

```python
#Menemukan data nan (null) menggunakan isna()
df.isna()
#keterangan : True menunjukkan bahwa data itu nan (null)
```

```python
df.isna().any()
```

```python
df.isna().sum()
#keterangan : jumlah data yg nan tiap kolom
```

```python
df.isnull().sum()
#keterangan : cara lain selain menggunakan isna()
```

```python
#mengubah karakter lain pada missing value menjadi nan (missing value diketahui)
missValue = ["?","--"]
df2 = pd.read_csv('newDF.csv', na_values=missValue)
df2
```

```python
df2.isna().sum()
```

```python
#cara lain untuk mengganti karakter pada missing value menjadi nan
df3 = df.replace({'?':np.nan, '--':np.nan})
df3
```

```python
df3.isna().sum()
```

#### Drop missing value

```python
df3
```

```python
df3.dropna(axis=0, how='all', inplace=True)
df3
```

```python
df3.dropna(axis=0, how='any')
#keterangan : menunjukkan nilai yang tidak ada missing value
```

```python
df3.dropna(axis=0, thresh=3)
```

#### Replacing missing value

```python
df3
```

```python
#Replacing dengan skalar
df4 = df3.fillna(0) #mengubah missing value menjadi nol
df4
```

```python
#Replacing dengan nilai rata2 (mean)
df3.iloc[:,0] = df3.iloc[:,0].fillna(df3.iloc[:,0].mean())
df3

#keterangan
#kolom target[semua baris:kolom 1]= kolom target.fillna(kolom target.mean())
```

```python
dfx = df.replace({'?':np.nan, '--':np.nan})
dfx
```

```python
#Replacing dengan nilai modus (nilai yang sering keluar)
dfx['col_A'].mode()
#keterangan : hasil mode ada 1 object dengan nilai 1.0
```

```python
dfx['col_A'] = dfx['col_A'].fillna(dfx['col_A'].mode()[0])
dfx
#keterangan :
#kolom target = kolom Target.fillna(kolom target.mode()[mode yg dipilih])
```

```python
df5 = df.replace({'?':np.nan, '--':np.nan})
df5
```

```python
#mengisi nilai nan dengan nilai sebelumnya
df5.fillna(method='ffill', inplace=True)
df5
```

```python
df6 = df.replace({'?':np.nan, '--':np.nan})
df6
```

```python
#mengisi nilai nan dengan nilai sesudahnya
df6.fillna(method='bfill', inplace=True)
df6
```
